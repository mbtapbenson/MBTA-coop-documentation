```python
# Change Types
voucher_df['VCH - Voucher Line No'] = voucher_df['VCH - Voucher Line No'].astype('str')
voucher_df['VCH - PO Line No'] = voucher_df['VCH - PO Line No'].astype('str')
voucher_df['VCH - PO Line Sched No'] = voucher_df['VCH - PO Line Sched No'].astype('str')
voucher_df['VCH - PO Distrib Line'] = voucher_df['VCH - PO Distrib Line'].astype('str')
voucher_df['VCH - Voucher Rcpt Line No'] = voucher_df['VCH - Voucher Rcpt Line No'].astype('str')

# Rename 
voucher_df.rename(columns={'Query Date': 'VCH - Query Date'}, inplace=True)
```

```python
voucher_df['Calc - Vchr Amt Movement'] = np.where(voucher_df['VCH - Voucher Line Amt'] < 0, 'Neg(-1)', 'Pos(+)')
voucher_df['VCH - Fund'] = voucher_df.apply(lambda x: str(x['VCH - Fund']).rjust(2, '0'), axis=1)
```
