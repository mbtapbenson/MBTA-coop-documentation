```python
class current_qoh_model:
    def __init__(self, base_list, item_dict):
        #print(base_list, item_dict)
        
        self.current_qoh_dict = {}
        # this will be a dict of base : item : QOH
        # item dict is a dict of base: item: quantity. item dict will be the starting QOH of January 1st, 2018.
        
        self.unique_items = set()
        self.unique_bases = set(base_list)
        
        for base in base_list:
            self.current_qoh_dict[base] = {}
            
            for item in item_dict[base]:
                self.current_qoh_dict[base][item] = item_dict[base][item]
                
                self.unique_items.add(item)

        self.current_orders = {}
        # This will be a dict of item: arrival date (at central), quantity
        
        self.item_base_stockout_dict = {}
        # base: item : number of stockouts
        self.item_central_stockout_dict = {}
        # item : days stocked out
        self.item_average_qoh = {}
        # base: item : [quantity on hand list]
        
        self.total_stockouts = {}
        # item : times when all bases don't have the item
        
        #print(self.unique_bases, self.unique_items)
        
    def run_simulation(self, transaction_dataframe, leadtime_df = leadtime_df, min_roq_df = full_part_df, 
                       start_date = datetime.datetime(2018, 1, 1), end_date = datetime.datetime(2022, 1, 1)):
        # january 1st, 2018
        self.current_date = start_date
        # End date: January 1st, 2022
        
        # Filter txn_df to just included items/bases
        filtered_txn_df = transaction_dataframe.loc[(transaction_dataframe['TXN - Item ID'].isin(self.unique_items)) & (transaction_dataframe['TXN - Unit'].isin(self.unique_bases))]
        
        #print(filtered_txn_df)
        
        self.txn_df = filtered_txn_df
        self.leadtime_df = leadtime_df
        self.min_roq_df = min_roq_df
        
        # Until the final entry in the transaction dataframe, we go through each day and track the transactions.
        day_delta = datetime.timedelta(days=1)
        while self.current_date <= end_date:
            
            print(self.current_date)
            
            # Filter to just daily transactions
            daily_transactions = self.txn_df.loc[self.txn_df['TXN - Transaction Date'] == self.current_date].to_dict('records')
            
            
            # Process orders here            
            self.process_part_orders()
            
            # append to average qoh list:
            for base_avg in self.current_qoh_dict:
                for item_avg in self.current_qoh_dict[base_avg]:
                    # Log average qoh
                    if base_avg in self.item_average_qoh:
                        if item_avg in self.item_average_qoh[base_avg]:
                            self.item_average_qoh[base_avg][item_avg].append(self.current_qoh_dict[base_avg][item_avg])
                        else:
                            self.item_average_qoh[base_avg][item_avg] = [self.current_qoh_dict[base_avg][item_avg]]
                    else:
                        self.item_average_qoh[base_avg] = {}
                        self.item_average_qoh[base_avg][item_avg] = [self.current_qoh_dict[base_avg][item_avg]]
            
            for item_cent in self.current_qoh_dict['CS004']:
                # Central stockouts/reorders
                if item_cent in self.current_qoh_dict['CS004']:
                    if self.current_qoh_dict['CS004'][item_cent] <= 0:
                        if item_cent in self.item_central_stockout_dict:
                            self.item_central_stockout_dict[item_cent] += 1
                        else:
                            self.item_central_stockout_dict[item_cent] = 1
                        
                        keys_without_central = list(self.current_qoh_dict.keys())
                        keys_without_central.remove('CS004')
                        
                        for base in keys_without_central:
                            if item_cent in self.current_qoh_dict[base] and self.current_qoh_dict[base][item_cent] <= 0:
                                if item_cent in self.total_stockouts:
                                    self.total_stockouts[item_cent] += 1
                                else:
                                    self.total_stockouts[item_cent] = 1

                    if self.current_qoh_dict['CS004'][item_cent] <= self.get_central_min(item_cent):
                        #print('Ordering more', self.current_qoh_dict['CS004'][item_avg], self.get_central_min(item_avg))
                        self.replenish_central(item_cent, self.current_date)

            for entry in daily_transactions:
                 # For each item: 
                # - Process usage
                # - calculate stockouts
                # - Process base replenishment 
                # - process central replenishment
                
                txn_item = entry['TXN - Item ID']
                txn_base = entry['TXN - Unit']
                txn_qty = entry['TXN - Qty']
                
                # process usage:
                self.use_part(txn_item, txn_base, txn_qty)
                
                # Calculate stockouts:
                for base in self.current_qoh_dict:
                    if base != 'CS004' and txn_item in self.current_qoh_dict[base]:
                        if self.current_qoh_dict[base][txn_item] <= 0:
                            
                            if base in self.item_base_stockout_dict:
                                if txn_item in self.item_base_stockout_dict[base]:
                                    self.item_base_stockout_dict[base][txn_item] += 1
                                else:
                                    self.item_base_stockout_dict[base][txn_item] = 1
                            else:
                                self.item_base_stockout_dict[base] = {}
                                self.item_base_stockout_dict[base][txn_item] = 1
                                
                        # Base replenishment:
                        if self.current_qoh_dict[base][txn_item] <= self.get_base_min(base, txn_item):
                            self.ship_to_base(txn_item, base)
            
            # go to the next day
            self.current_date += day_delta
            
    def get_central_min(self, item):
        #print(item)
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item)].head(1)['central_min'].item()

    def get_central_roq(self, item):
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item)].head(1)['central_roq'].item()

    def get_base_min(self, base, item):
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item) & (self.min_roq_df['Base'] == base)]['base_min'].item()
    
    def get_base_roq(self, base, item):
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item) & (self.min_roq_df['Base'] == base)]['base_roq'].item()
        
    def ship_to_base(self, item, base):
        # Ships quantity parts from central store to the specified base
        roq = self.get_base_roq(base, item)
        
        if self.current_qoh_dict['CS004'][item] > roq:
            self.current_qoh_dict['CS004'][item] -= roq
            self.current_qoh_dict[base][item] += roq
        else:
            self.current_qoh_dict[base][item] += self.current_qoh_dict['CS004'][item]
            self.current_qoh_dict['CS004'][item] = 0
            

    def replenish_central(self, item, purch_date):
        roq = self.get_central_roq(item)
        
        # Replenishes central as though it's a purchase through the buyers
        leadtime_days = datetime.timedelta(days = self.get_most_recent_leadtime(item, purch_date))
        
        if item not in self.current_orders or self.current_orders[item] == -1:
            self.current_orders[item] = {'arrival_date': purch_date + leadtime_days,
                                        'quantity' : roq}
    
    def process_part_orders(self):
        # If current_date == arrival date: replenish item with quantity, clear entry from dict
        
        for item in self.current_orders:
            if self.current_orders[item] != -1 and self.current_date == self.current_orders[item]['arrival_date']:
                self.current_qoh_dict['CS004'][item] += self.current_orders[item]['quantity']
                
                self.current_orders[item] = -1


    def get_most_recent_leadtime(self, item, min_reach_date):
        # Gets the leadtime to use for replenishment 
        # min_reach_date is when the item hit the min
        
        # Filter leadtime df to just this item, and all transactions before min_reach_date
        
        temp_lt_df = self.leadtime_df.loc[(self.leadtime_df['TXN - Item ID'] == item) & (self.leadtime_df['PO Date'] < min_reach_date)]
        
        if len(list(temp_lt_df.sort_values(by='PO Date', ascending = False)['Lead Time'])) > 0:
            leadtime_to_use = list(temp_lt_df.sort_values(by='PO Date', ascending = False)['Lead Time'])[0]
        else:
            # heuristic: if leadtime isn't here, just use a month.
            leadtime_to_use = 30
        
        return leadtime_to_use
    
    def use_part(self, item, base, quantity):
        # uses an item at a base
        if item in self.current_qoh_dict[base]:
            self.current_qoh_dict[base][item] -= quantity

            if self.current_qoh_dict[base][item] <= 0:
                self.current_qoh_dict[base][item] = 0

    def __repr__(self):
        # returns a string
        
        final_string = ''
        final_string += 'Current QOH:\n' + str(self.current_qoh_dict)
        
        final_string += '\nAverage QOH:\n'
        
        for base in self.item_average_qoh:
            for item in self.item_average_qoh[base]:
                average_item = str(sum(self.item_average_qoh[base][item]) / len(self.item_average_qoh[base][item]))
                final_string += 'Item: {}; Base: {} '.format(item, base) + average_item + '\n'
 
        final_string += '\nBase Stockouts:\n' + str(self.item_base_stockout_dict)
        final_string += '\nCentral Stockouts:\n' + str(self.item_central_stockout_dict)
        final_string += '\nComplete Stockouts:\n' + str(self.total_stockouts)
        
        return final_string
        
```

```python
import pandas as pd
import numpy as np
import os
import datetime
import bisect

class item_current_qoh_model:
    def __init__(self, item_dict) -> None:
        self.current_qoh_dict = item_dict
        # Item : base : qoh

        # Average QOH
        self.average_qoh = {}
        # item : average qoh
        # calculated by summing, then dividing at the end of the simulation by how many days passed. 
        
        # Stockouts at Central:
        self.central_stockouts = {}
        # item : stockouts

        print('Created Model.')

    def run_simulation(self, full_transaction_df, leadtime_df, min_roq_df, start_date, end_date):
        # Don't iterate thourgh each day. Run each item separately. 
        self.min_roq_df = min_roq_df
        
        self.min_roq_df['unit_cost'] = self.min_roq_df['Qty On Hand'] / self.min_roq_df['On Hand Value']
        
        self.leadtime_df = leadtime_df

        # Rename full_transaction_df columns
        renamed_txn_df = full_transaction_df.rename(columns={'TXN - Unit': 'Base', 'TXN - Qty':'Qty', 'TXN - Transaction Date':'Date'})

        #print(renamed_txn_df)
            
        # Create events list for each item.
        for item in self.current_qoh_dict:
            # Create event list, then iterate through it

            item_event_list = self.create_event_list(item, renamed_txn_df, start_date, end_date)
            # leadtime events are added later
            #print(item_event_list)

            self.run_item_simulation(item, item_event_list)

    
    def run_item_simulation(self, item, item_event_list):
        # This kind of iteration will probably through a bug 
        i = 0
        while i < len(item_event_list):
            
            
            # Calculate average QoH
            if i > 0:
                days_difference = item_event_list[i]['Date'] - item_event_list[i-1]['Date']
                
                if days_difference.days > 0:
                    if item in self.average_qoh:
                        self.average_qoh[item] += round(sum(self.current_qoh_dict[item].values()) / days_difference.days)
                    else:
                        self.average_qoh[item] = round(sum(self.current_qoh_dict[item].values()) / days_difference.days)
            
            item_event_to_process = item_event_list[i]
            temp_length = len(item_event_list)
    
            self.process_item_event(item, item_event_to_process, item_event_list) 
            # This function is adding a duplicate to the item event list
            
            #print(i, item_event_list[i], len(item_event_list))
            #print('\n')
            #print(item_event_list)
            
            i += 1

    def create_event_list(self, item, txn_df, start_date, end_date):
        # Gets only the transactions for the specified item, in the right timeframe.
        temp_txn_df = txn_df.loc[(txn_df['TXN - Item ID'] == item) & (txn_df['Date'] >= start_date) & (txn_df['Date'] <= end_date)]

        # Convert to records, sort by transaction date
        final_list = sorted(temp_txn_df.to_dict('records'), key=lambda txn:txn['Date'])
        # This is a dict of out transactions, which includes Item ID, Base, Quantity (out qty), and date

        # Go through final_list and flip the sign of the quantity
        for entry in final_list:
            entry['Qty'] = -1 * entry['Qty']

        return final_list
    
    def process_item_event(self, item, item_event, item_event_list):
        if item_event['Base'] in self.current_qoh_dict[item]:
            self.current_qoh_dict[item][item_event['Base']] += item_event['Qty']

            # Check for stockouts, check for reorders
            # Also check for base replenishment
            if self.current_qoh_dict[item][item_event['Base']] < 0:
                self.current_qoh_dict[item][item_event['Base']] = 0


            if item_event['Base'] != 'CS004':
                if self.current_qoh_dict[item][item_event['Base']] <= self.get_base_min(item_event['Base'], item):
                    #print('Base reorder at: ')
                    #print(self.current_qoh_dict[item][item_event['Base']])
                    # reorder from base 
                    self.base_transfer(item_event['Base'], item, self.get_base_roq(item_event['Base'], item), item_event['Date'], item_event_list)
            else:
                if self.current_qoh_dict[item]['CS004'] <= self.get_central_min(item):
                    self.order_parts(item, item_event['Date'], item_event_list)
        
    def add_item_event(self, item_event, item_event_list):
        # Add the item event in the proper place
        for i in range(len(item_event_list)):
            if item_event_list[i]['Date'] > item_event['Date']:
                item_event_list.insert(i, item_event)
                break

    def get_most_recent_leadtime(self, item, date):
        # Gets the leadtime to use for replenishment 
        
        # Filter leadtime df to just this item, and all transactions before min_reach_date
        
        temp_lt_df = self.leadtime_df.loc[(self.leadtime_df['TXN - Item ID'] == item) & (self.leadtime_df['PO Date'] < date)]
        
        if len(list(temp_lt_df.sort_values(by='PO Date', ascending = False)['Lead Time'])) > 0:
            leadtime_to_use = list(temp_lt_df.sort_values(by='PO Date', ascending = False)['Lead Time'])[0]
        else:
            # heuristic: if leadtime isn't here, just use a month.
            leadtime_to_use = 30
        
        return leadtime_to_use
    
    def order_parts(self, item, date, item_event_list):
        leadtime = datetime.timedelta(days = self.get_most_recent_leadtime(item, date))

        # Insert into the item event list (inplace?)
        # Insert so that the item is sorted properly
        item_order_event = {'TXN - Item ID' : item,
                            'Base': 'CS004',
                            'Date': date + leadtime,
                            'Qty': self.get_central_roq(item)}

        # Add the order to the list of events
        #print('Item Order: ', item_order_event)
        self.add_item_event(item_order_event, item_event_list)
    
    def base_transfer(self, base, item, qty, date, item_event_list):
        # instantaneous base transfer

        if self.current_qoh_dict[item]['CS004'] <= qty:
            self.current_qoh_dict[item][base] += self.current_qoh_dict[item]['CS004']
            self.current_qoh_dict[item]['CS004'] = 0
            
            # log stockouts
            if item in self.central_stockouts:
                self.central_stockouts[item] += 1
            else:
                self.central_stockouts[item] = 1
        else:
            self.current_qoh_dict[item][base] += qty
            self.current_qoh_dict[item]['CS004'] -= qty

        if self.current_qoh_dict[item]['CS004'] <= self.get_central_min(item):
            # Trigger central reorder event
            self.order_parts(item, date, item_event_list)

    # Get min and roq for base/items
    def get_central_min(self, item):
        #print(item)
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item)].head(1)['central_min'].item()

    def get_central_roq(self, item):
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item)].head(1)['central_roq'].item()

    def get_base_min(self, base, item):
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item) & (self.min_roq_df['Base'] == base)]['base_min'].item()
    
    def get_base_roq(self, base, item):
        return self.min_roq_df.loc[(self.min_roq_df['TXN - Item ID'] == item) & (self.min_roq_df['Base'] == base)]['base_roq'].item()
        
    def get_item_cost(self, item):
        #print(self.min_roq_df.columns)
        #print(self.min_roq_df)
        cost = self.min_roq_df.loc[self.min_roq_df['TXN - Item ID'] == item]['unit_cost'].iloc[0]
        
        if cost != cost:
            return 0
        else:
            return cost
    
    def __repr__(self) -> str:
        final_string = '\n'
        final_string += 'Average QOH:\n' + str(self.average_qoh)
        final_string += '\nCentral Stockouts:\n' + str(self.central_stockouts)
        
        # Calculate cost of average QOH:
        total_cost = 0
        for item in self.average_qoh:
            total_cost += self.get_item_cost(item) * self.average_qoh[item]
        
        final_string += '\nTotal Cost of Average QOH: ' + str(total_cost)
        
        final_string += '\nTotal Stockouts: ' + str(sum(self.central_stockouts.values())) + '\n'
        
        return final_string
```
