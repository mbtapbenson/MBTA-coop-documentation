```python

safety = 2
# Higher safety means fewer stockouts and more on hand at a time. Safety of 2 should account for 99.7% of situations.
quarters_per_purchase = 2
# How often we buy. Right now, we are using a heuristic of "buyers buy once-twice per year."

full_base_part_df['part_minimum'] = full_base_part_df['leadtime_mean'] * (full_base_part_df['usage_mean'] \
                                                                          + (safety * full_base_part_df['usage_std']))
full_base_part_df['part_roq'] = (full_base_part_df['leadtime_mean'] * full_base_part_df['usage_mean']) \
                                                                          + (quarters_per_purchase * full_base_part_df['usage_mean'])
    
```

```python
txn_converter = {'TXN - Item ID': str}
central_quarters_df = pd.read_csv('Data/central_store_quarters_df.csv', index_col = 0, converters=txn_converter)
central_parts_df = pd.read_csv('Data/central_store_parts_df.csv', index_col=0, converters=txn_converter)
```

```python
def calc_min_roq(part_df, safety = 1, quarters_per_purchase = 2):
    copied_df = part_df.copy(deep=True)

    copied_df['part_minimum'] = copied_df['leadtime_mean'] * (copied_df['usage_mean'] + (safety * copied_df['usage_std']))
    copied_df['part_roq'] = (copied_df['leadtime_mean'] * copied_df['usage_mean']) + (quarters_per_purchase * copied_df['usage_mean'])
    
    return copied_df
```

```python
def calc_base_min_roq(base, part_number, base_part_df, central_mins_df):
    # get central part min/roq
    central_min = central_mins_df.loc[central_mins_df['TXN - Item ID'] == part_number]['part_minimum']
    central_roq = central_mins_df.loc[central_mins_df['TXN - Item ID'] == part_number]['part_roq']
    
    # calculate base propotional usage
    total_base_usage = base_part_df.groupby('TXN - Item ID')['usage_mean'].sum()[part_number]
    
    individual_base_usage = base_part_df.loc[(base_part_df['TXN - Item ID'] == part_number) & (base_part_df['Base'] == base)]['usage_mean'].item()
    
    base_factor = individual_base_usage / total_base_usage
    
    if len(central_min) == 0 and len(central_roq) == 0:
        central_min = 2 * total_base_usage
        central_roq = 4 * total_base_usage
    else:
        central_min = central_min.item()
        central_roq = central_roq.item()
    
    individual_base_min = 0.5 * central_min * base_factor
    individual_base_roq = 0.5 * central_roq * base_factor
    
    return {'TXN - Item ID':part_number, 'Base':base, 'base_min': individual_base_min, 'base_roq': individual_base_roq}
```

```python
central_with_mins = calc_min_roq(central_parts_df, safety=2)
central_with_mins
```

```python
def calc_quarters(txn_df):
    # needs a dataframe with 'TXN - Transaction Date'
    months_df = txn_df.copy(deep=True)

    start_date = datetime.datetime(2004, 1, 1)

    # Calculate quarters
    months_df['Quarters'] = ((months_df['TXN - Transaction Date'] - start_date).astype('timedelta64[M]').astype('int') - 1) // 3
    
    return months_df
    
```

```python
def get_all_out_parts(txn_df):
# get active parts df here
    today = datetime.date.today()

    out_codes = ('030', '031')
    # Filter to at least 5 usages in the last 2 years. 
    last_2_years = txn_df.copy(deep=True)

    last_2_years = last_2_years.loc[last_2_years['TXN - Transaction Type'].isin(out_codes)]
    #last_2_years = last_2_years.loc[last_2_years['TXN - Transaction Type'] == '030']

    last_2_years = last_2_years[['TXN - Transaction Date', 'TXN - Transaction Type', 'TXN - Item ID']]

    # This is not the right understanding – this eliminates all rows that are at least 2 years old. 
    # We want to remove all parts that haven't been used in 2 years. 
    last_2_years['Today'] = today
    last_2_years['Today'] = pd.to_datetime(last_2_years['Today'])
    last_2_years['TXN - Transaction Date'] = pd.to_datetime(last_2_years['TXN - Transaction Date'])
    last_2_years['Years Ago'] = (last_2_years['TXN - Transaction Date'] - last_2_years['Today']).astype('timedelta64[Y]').astype('int')

    # Need to find the usage in the past 2 years.
    last_2_years_df = last_2_years.copy(deep=True)

    # Only filter out parts if the option is on.
    last_2_years_df = last_2_years_df.loc[last_2_years_df['Years Ago'] >= -2]

    print(last_2_years_df['Years Ago'].min())

    part_usages = last_2_years_df.groupby('TXN - Item ID').size()
    part_filter = part_usages > 5
    in_parts = set(part_usages[part_filter].index)
    
    return(in_parts)
```

# Visuals for Takary

```python
def simulate_use(leadtime, mean, std, quarters_per_purch, total_quarters, resolution):
    item_min = round(leadtime * (mean + (2 * std)))
    item_roq = round(item_min + (quarters_per_purch * mean))
    
    num_points = int(1/resolution) * total_quarters
    
    quantities = []
    quantities.append(item_min + item_roq)
    
    usages = np.random.normal(loc = mean, scale=std, size=num_points) 
    usages = usages * resolution
    #print(usages)
    
    add_roq = -1
    
    for i in range(1, num_points):
        
        #print(i)
        quantities.append(quantities[i-1] - round(usages[i]))
        
        #print(i)
        if i == add_roq:
            #print('reorder arrived')
            #print(quantities[i])
            quantities[i] += item_roq
            add_roq = -1
        
        if quantities[i] <= item_min and add_roq == -1:
            add_roq = i + int(int(1/ resolution) * leadtime)
            
            if add_roq == i:
                add_roq += 1
            #print('add_roq:', add_roq)
            
    #print(1/resolution)
    
    quarter_list = np.arange(total_quarters)
    #print(quarter_list * 6.25)
    
    figure(figsize=(10, 5), dpi=80)
    
    fig = plt.figure(figsize=(20,5))
    ax = fig.add_subplot()
    ax.plot(list(range(num_points)), quantities)
    ax.hlines(y=item_min, xmin=0, xmax=num_points, color='r', linestyle='-')
    ax.vlines(x=quarter_list * (1/resolution), ymin=0, ymax=max(quantities), color='b', linestyle='--', alpha=0.1)
    ax.set_title('Usage Simulation \nMean: {}, Leadtime: {}, STD: {}, Quarters per Purchase: {} \nMin: {} ROQ: {}'.format(mean, leadtime, std, quarters_per_purch, item_min, item_roq))
    ax.set_ylabel('Quantity on Hand')
    ax.set_xlabel('Time')
    
    #print(quantities)
```

```python
# LT, mean, std, quarters per purch, quarters to run, resolution
simulate_use(0.5, 10, 2, 2, 20, 0.16)
```
