```python
def visualize_part(part_df, quarter_df, part_number):
    
    example_part_df = part_df.loc[part_df['TXN - Item ID'] == part_number]
    example_quarter_df = quarter_df.loc[quarter_df['TXN - Item ID'] == part_number]
    
    new_reord_pts = calc_min_roq(example_part_df, part_number)
    #print(new_reord_pts)

    fig = plt.figure()
    ax = fig.add_subplot()
    ax.set_ylabel('Qty')
    ax.set_xlabel('Quarter since 2004')
    ax.set_title('Part number {}'.format(part_number) + '\n Qty STD: {:.2f} Mean Quarterly Usage: {:.2f} Mean Leadtime: {:.2f} quarters'.format(new_reord_pts['usage_std'], new_reord_pts['usage_mean'], new_reord_pts['lead_time']) 
                 + '\n Min: {:.2f} ROQ: {:.2f}'.format(new_reord_pts['min'], new_reord_pts['roq']))
    ax.scatter(example_quarter_df['Quarters'], example_quarter_df['TXN - Qty'], color = 'darkblue', s=25)
    
    # Add standard deviation horizontal lines
    ax.hlines(y=[new_reord_pts['min'], new_reord_pts['roq']], xmin=0, xmax=max(example_quarter_df['Quarters']), colors=['red', 'purple', 'blue'], linestyles='--', lw=1)    
    
    plt.show
```

```python
def calc_min_roq(part_df, part_number, recency = 60, aggression = 1, cost_factor=2):
    # Filter to the part row 
    
    part_row = part_df.loc[part_df['TXN - Item ID'] == part_number]
    #print(part_row)
    
    usage_mean = list(part_row['usage_mean'])[0]
    usage_std = list(part_row['usage_std'])[0]
    lead_time_quarters = list(part_row['leadtime_mean'])[0]
    lead_time_std = list(part_row['leadtime_std'])[0]
    
    part_minimum = lead_time_quarters * (usage_mean + (aggression * usage_std))
    part_roq = (lead_time_quarters * usage_mean) + (cost_factor * usage_mean)
    
    return {'part_number':part_number, 'usage_mean':usage_mean, 'lead_time':lead_time_quarters, 'lead_time_std':lead_time_std, 'usage_std':usage_std, 'min':part_minimum, 'roq':part_roq}
```

```python
# Here, we calculate the min and roq for bases. 

# Higher safety means fewer stockouts and more on hand at a time. Safety of 2 should account for 99.7% of situations.
# How often we buy. Right now, we are using a heuristic of "buyers buy once-twice per year."

def calc_min_roq(part_df, safety = 1, quarters_per_purchase = 2):
    copied_df = part_df.copy(deep=True)

    copied_df['part_minimum'] = copied_df['leadtime_mean'] * (copied_df['usage_mean'] + (safety * copied_df['usage_std']))
    copied_df['part_roq'] = (copied_df['leadtime_mean'] * copied_df['usage_mean']) + (quarters_per_purchase * copied_df['usage_mean'])
    
    return copied_df
```

Total base mins = 0.5 * central mins

Individual base mins = proportional usage * total base mins. 

Example: if BS018 uses part #0002 twice per Q and BS009 uses part #0002 four times per Q, BS009 would have twice the min as BS018. If this is between 0 and 1, it's rounded to 1. 

```python
def calc_base_min_roq(base, part_number, base_part_df, central_mins_df):
    # get central part min/roq
    central_min = central_mins_df.loc[central_mins_df['TXN - Item ID'] == part_number]['part_minimum']
    central_roq = central_mins_df.loc[central_mins_df['TXN - Item ID'] == part_number]['part_roq']
    
    # calculate base propotional usage
    total_base_usage = base_part_df.groupby('TXN - Item ID')['usage_mean'].sum()[part_number]
    
    individual_base_usage = base_part_df.loc[(base_part_df['TXN - Item ID'] == part_number) & (base_part_df['Base'] == base)]['usage_mean'].item()
    base_factor = individual_base_usage / total_base_usage
    
    if len(central_min) == 0 and len(central_roq) == 0:
        central_min = 2 * total_base_usage
        central_roq = 4 * total_base_usage
    else:
        central_min = central_min.item()
        central_roq = central_roq.item()
    
    individual_base_min = 0.5 * central_min * base_factor
    individual_base_roq = 0.5 * central_roq * base_factor
    
    return {'TXN - Item ID':part_number, 'Base':base, 'base_min': individual_base_min, 'base_roq': individual_base_roq}
```

# Scheduling Cost Savings

If we purchase a part yearly, we must buy enough for that year. This means that, to be safe, we need 4 * (mean monthly usage + (2 * std)) parts per year. We must then store all of these parts until they are all used up. If we want to be safe on a "quarterly" basis, we need enough parts for that quarter.
