```python
import pandas as pd
import numpy as np
import os
import datetime

import matplotlib.pyplot as plt

os.chdir('/Users/pbenson/Documents/Min_Max_Project')
```

```python
stockout_df = pd.read_csv('Data/Stockouts/WH_STOCKOUTS_FULL.csv')
```

```python
usage_df = pd.read_csv('Data/Stockouts/full_usage_df.csv')
```

```python
# Filter stockouts, label the filterings
```

```python
filtered_stockouts = stockout_df[['Item', 'Item Group', 'Status Current', 'No Repl', 'Reorder Pt', 'Max Qty', 'Reord Qty', 'Trans Date']]
```

```python
filtered_stockouts['Item'] = filtered_stockouts['Item'].astype('str')
```

```python
stockouts_with_usage = filtered_stockouts.merge(usage_df, how='left', left_on='Item', right_on='TXN - Item ID')

stockouts_with_usage.drop('TXN - Item ID', axis=1, inplace=True)

stockouts_with_usage = stockouts_with_usage.fillna(0)
```

```python


def stockout_filter(row, usage_column, exclude_groups):    
    # Filter out by Status, No Repl, and Reorder stuff
    if (row['No Repl'] == 'Y'):
        return False
    
    elif (row['Status Current'] == 'Discontinue'):
        return False
    
    elif (row['Reorder Pt'] == 0 and row['Max Qty'] == 0 and row['Reord Qty'] == 0):
        return False
    
    else:
        if row[usage_column] == 0:
            return False
        
        else:
            if row['Item Group'] in exclude_groups:
                return False
            else:
                return True

def stockout_reason(row, usage_column, exclude_groups):    
    # Filter out by Status, No Repl, and Reorder stuff
    if (row['No Repl'] == 'Y'):
        return 'no_repl_count'
    
    elif (row['Status Current'] == 'Discontinue'):
        return 'status_count'
    
    elif (row['Reorder Pt'] == 0 and row['Max Qty'] == 0 and row['Reord Qty'] == 0):
        return 'max_count'
    
    else:
        if row[usage_column] == 0:
            return 'usage_count'
        
        else:
            if row['Item Group'] in exclude_groups:
                return 'exclude_count'
            else:
                return 'none'
            
    # Filter out by usage.
    # Usage column = which column to check for usage. 
    
def filter_stockouts(stockout_df, usage_column, exclude_groups):
    temp_stock_df = stockout_df.copy(deep=True)
    
    reason_dict = {'no_repl_count': 0,
                   'status_count': 0, 
                   'max_count': 0, 
                   'usage_count': 0, 
                   'exclude_count': 0, 
                   'none': 0}
        
    temp_stock_df['filter'] = temp_stock_df.apply(stockout_filter, args=(usage_column, exclude_groups), axis=1)
    temp_stock_df['reason'] = temp_stock_df.apply(stockout_reason, args=(usage_column, exclude_groups), axis=1)
    
    for r in temp_stock_df['reason']:
        reason_dict[r] += 1
    
    filtered_df = temp_stock_df.loc[temp_stock_df['filter'] == True]
    filtered_df.drop(['filter', 'reason'], axis=1, inplace=True)
    
    print(reason_dict)
    
    return filtered_df
    
```

```python

final_stockouts = filter_stockouts(stockouts_with_usage, 'usage_5_years', ['00'])
```

Stockout metric: number of stockouts per quarter and unique items that experienced stockouts in that quarter

```python
# Groupby Quarter

final_stockouts['Trans Date'] = pd.to_datetime(final_stockouts['Trans Date'])

final_stockouts['epoch'] = datetime.datetime(2004, 1, 1)

final_stockouts['months_since_2004'] = (final_stockouts['Trans Date'].dt.to_period('M').astype(int) - final_stockouts['epoch'].dt.to_period('M').astype(int))

final_stockouts['quarter'] = final_stockouts['months_since_2004'] // 4
final_stockouts['year'] = final_stockouts['months_since_2004'] // 12
```

```python
final_stockouts
```

```python
stockouts_per_quarter = final_stockouts.groupby('quarter').size()
stockouts_per_year = final_stockouts.groupby('year').size()
```

```python
print(stockouts_per_quarter)
print(stockouts_per_year)
```

```python
unique_stockouts_per_quarter = final_stockouts.groupby('quarter')['Item'].nunique()
unique_stockouts_per_year = final_stockouts.groupby('year')['Item'].nunique()
```

```python
print(unique_stockouts_per_quarter)
print(unique_stockouts_per_year)
```

```python
stockout_metrics = pd.DataFrame({'stockouts': stockouts_per_quarter, 'unique_stockouts': unique_stockouts_per_quarter})
stockout_metrics['proportion'] = stockout_metrics['stockouts'] / stockout_metrics['unique_stockouts']
```

```python
stockout_metrics
```

```python
plt.plot(stockout_metrics['proportion'])
```
